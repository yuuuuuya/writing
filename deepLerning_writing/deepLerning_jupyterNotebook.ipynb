```python
import numpy as np
import matplotlib.pylab as plt

def sigmoid(x):
  return 1 / (1+np.exp(-x))

x = np.arange(-50.0, 50.0, 0.1)
y = sigmoid(x)

plt.title("sigmoid function")
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
import numpy as np
import matplotlib.pylab as plt

def relu(x):
  return np.maximum(0, x)

x = np.arange(-50.0, 50.0, 0.1)
y = relu(x)
plt.title('ReLU function')
plt.plot(x, y)
plt.show()
```
